```python
import matplotlib.pyplot as plt
# Plotting class imbalance
plt.figure(figsize=(6,4))
plt.bar(data['Churn'].value_counts().index, data['Churn'].value_counts().values)
plt.title("Churn vs No Churn Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(data['Contract'].value_counts().index, data['Contract'].value_counts().values)
plt.title("Contract Type vs Churn")
plt.xlabel('contract type')
plt.ylabel('count of churned customers')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(data['InternetService'].value_counts().index, data['InternetService'].value_counts().values)
plt.xlabel('Internet Service Type')
plt.ylabel('Churn')
plt.title("Internet Service vs Churn")
plt.show()
```
